```python
import csv
import math as m
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def file(data_path):
    with open(data_path, 'r') as inputfile:
        reader = csv.reader(inputfile, delimiter=';')
        headers = next(reader)
    with open (data_path,'r') as inputfile:
        data=np.genfromtxt(inputfile, delimiter=";", skip_header=1)
    x=data[:,:-1] 
    y=data[:,-1]
    X=np.array(x,dtype='float')
    Y=np.array(y,dtype='int')
    H=np.array(headers,dtype=object)
    return X,Y,H

def histogram_bins(C,bin_size):
    N=len(C)
    bin_size = round(2*(N)**(1/3))+1 ##(using Rice's rule)
    min_data_set=np.min(C)
    max_data_set=np.max(C)
    bins=np.linspace(min_data_set,max_data_set,bin_size)
    return bins,bin_size
    
path='D:\STUDY MATERIAL\SECOND SEM\MACHINE LEARNING\EXERCISE\winequality-white.csv'
#A=Attributes
#x=element parameter
#y=quality index vector
x,y,A=file(path)
```

```python
fig,axes = plt.subplots(3,figsize=(10, 25))
fig.suptitle("symmetric  vs unsymmetric", fontsize=16)
i=8
C=x[:,i]  
X=np.ones_like(C)/4898
count,bin_size=np.histogram(C,bins=20,weights=X)
CP=round(max(np.diff(bin_size)),3)
print("BIN SIZE FOR PH VALUE:",CP)
axes[0].hist(C,bins=20,weights=X,stacked=True,color="green",rwidth=0.90)
axes[0].set_xlabel(A[i])
axes[0].set_ylabel("probability")
axes[0].set_title('probability of PH value')

i=9
C=x[:,i]  
count,bin_size=np.histogram(C,bins=20,weights=X)
CP=round(max(np.diff(bin_size)),3)
print("BIN SIZE FOR sulphates:",CP)
axes[1].hist(C,bins=20,weights=X,stacked=True,color="blue",rwidth=0.90)
axes[1].set_xlabel(A[i])
axes[1].set_ylabel("probability")
axes[1].set_title('probability of sulphates')

i=3
C=x[:,i]  
count,bin_size=np.histogram(C,bins=20,weights=X)
CP=round(max(np.diff(bin_size)),3)
print("BIN SIZE FOR residual sugar:",CP)
axes[2].hist(C,bins=20,weights=X,stacked=True,color="black",rwidth=0.90)
axes[2].set_xlabel(A[i])
axes[2].set_ylabel("probability")
axes[2].set_title('probability of residual sugar')

plt.show()
```
